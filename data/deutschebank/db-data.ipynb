```python
import pandas as pd
```

```python
cols = ["city_ascii", "country", "region", "price", "latitude", "longitude", "population"]
```

```python
def db_clean(filename):
    df = pd.read_csv(filename)
    df = df.rename(columns={'canonical city': "city_ascii"})
    df['region'] = df['region'].str.title()
    df = df.reindex(columns=cols)
    
    return df
```

```python
db_beer = db_clean("beer-pub.csv")
db_beer.to_csv('beer-clean.csv')
```

```python
db_cafe = db_clean("cappuccino.csv")
db_cafe.to_csv("cappuccino-clean.csv")
```
