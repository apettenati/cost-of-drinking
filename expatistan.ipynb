```python
import pandas as pd
import json
with open('data/expatistan/expatistan.json') as f:
    data = json.load(f)

expatistan = pd.json_normalize(data)
expatistan.columns = ['city', 'beer_supermarket', 'cappuccino', 'beer_pub']
expatistan['beer_supermarket'] = expatistan['beer_supermarket'].astype(float)
expatistan['cappuccino'] = expatistan['cappuccino'].astype(float)
expatistan['beer_pub'] = expatistan['beer_pub'].astype(float)
expatistan = expatistan.rename(columns = {"city": "city_ascii"})
# expatistan = expatistan.set_index('city_ascii')
```
