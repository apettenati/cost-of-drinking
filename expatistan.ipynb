```python
import pandas as pd
import json
with open('data/expatistan/expatistan-data.json') as f:
    data = json.load(f)

expatistan = pd.json_normalize(data)
expatistan.columns = ['city_ascii', 'country', 'beer_supermarket', 'bread', 'cappuccino', 'beer_pub']
expatistan['data_source'] = 'expatistan'
expatistan.set_index('city_ascii')
```

```python
expatistan.describe()
```

```python
cities = pd.read_csv('data/city-data/world-cities-data.csv')
cities.set_index('city_ascii')
```

```python
# adding city data - guessing we won't need this df if we're merging all data sources together?
new_df = pd.merge(expatistan, cities[['city_ascii', 'country', 'region', 'latitude', 'longitude', 'population']], how='left', on=['city_ascii', 'country'])
new_df
```
